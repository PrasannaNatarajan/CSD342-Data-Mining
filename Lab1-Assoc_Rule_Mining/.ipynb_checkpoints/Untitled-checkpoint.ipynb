```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

words = ' '.join(data_sorted['Item Sets'])

print words
wordcloud = WordCloud(
                      stopwords=STOPWORDS,
                      background_color='black',
                     ).generate(words)

plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('./my_first_wordcloud_1.png', dpi=300)
plt.show()
```
